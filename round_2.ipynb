```python
import pandas as pd
import plotnine as p9
```

```python
round_2_df = pd.read_csv("data_round_2.csv", sep=';')
round_2_df.columns
```

```python
squid_df = round_2_df[round_2_df['product'] == "SQUID_INK"]
basket1_df = round_2_df[round_2_df['product'] == "PICNIC_BASKET1"]
basket2_df = round_2_df[round_2_df['product'] == "PICNIC_BASKET2"]
resin2_df = round_2_df[round_2_df['product'] == "RAINFOREST_RESIN"]
kelp_df = round_2_df[round_2_df['product'] == "KELP"]
croissants_df = round_2_df[round_2_df['product'] == "CROISSANTS"]
jam_df = round_2_df[round_2_df['product'] == "JAMS"]
djembes_df = round_2_df[round_2_df['product'] == "DJEMBES"]
```

```python
(p9.ggplot(djembes_df, p9.aes(x='timestamp', y='mid_price', color='product')) +
    p9.geom_line() +
    p9.labs(title='Mid Price Over Time by Product', 
            x='Timestamp', 
            y='Mid Price',
            color='Product') +
    p9.theme_minimal() +
    p9.theme(figure_size=(12, 8)))
```
